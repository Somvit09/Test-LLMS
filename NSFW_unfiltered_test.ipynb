```
from diffusers import DiffusionPipeline
import torch
```

```
if torch.cuda.is_available():
    print("CUDA is available.")
    print(f"Current device: {torch.cuda.current_device()}")
    print(f"Device name: {torch.cuda.get_device_name(torch.cuda.current_device())}")
else:
    print("CUDA is not available.")
```

```
pipeline = DiffusionPipeline.from_pretrained("UnfilteredAI/NSFW-gen-v2")
```

```
pipeline = pipeline.to("cuda").to(torch.float16)
```

```
pipeline.enable_attention_slicing()
pipeline.enable_sequential_cpu_offload()
```

```
import os


# Generate the image
def generate_image(image_prompt: str, image_name: str):
  image = pipeline(image_prompt).images[0]
  # Use the current working directory for interactive environments
  BASE_DIR = os.getcwd()
  IMAGE_PATH = os.path.join(BASE_DIR, "images")

  # Create the "images" directory if it doesn't exist
  os.makedirs(IMAGE_PATH, exist_ok=True)

  # Save the image
  image.save(os.path.join(IMAGE_PATH, f"{image_name}.png"))
  return image
```

```
image = generate_image("earth in 2100 year.", "earth_2100")
image.show()
```
